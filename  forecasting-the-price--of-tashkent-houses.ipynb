```python
# Foydalanish imkonsiz bo'lgan location ustunini olib tashlash va alohida DataFramega saqlash

df_processed = df.drop("location", axis=1)
df_processed.head(5)
```

```python
# size ustunida int64da aylantirilib bo'lmaydigan qiymatlarni tekshirish

df_processed[df_processed['size'] == 'Площадьземли:1сот']
```

```python
# O'sha qatorni olib tashlash

df_processed.drop(df_processed[df_processed['size'] == 'Площадьземли:1сот'].index, inplace = True)
```

```python
# Qayta tekshirib ko'rish

df_processed[df_processed['size'] == 'Площадьземли:1сот']
```

```python
# size ustunini float64ga aylantirish

df_processed["size"] = df_processed["size"].astype(object).astype(float)
```

```python
# size ustunini int64ga aylantirish

df_processed["size"] = df_processed["size"].astype(float).astype(int)
```

```python
# price usutidagi son bo'lmagan qiymatlarni tekshirish

df_processed[df_processed['price'] == 'Договорная']
```

```python
# pricedagi son bo'lmagan qiymatlarni tashlab yuborish va o'ziga saqlab qo'yish

df_processed.drop(df_processed[df_processed['price'] == 'Договорная'].index, inplace = True)
```

```python
# price ustunini tekshirib ko'rish

df_processed[df_processed['price'] == 'Договорная']
```

```python
# price ustunini float64ga aylantirish

df_processed["price"] = df_processed["price"].astype(object).astype(float)
```

```python
# price ustunini int64ga aylantirish

df_processed["price"] = df_processed["price"].astype(float).astype(int)
```

```python
# Tayyor bo'lgan DataFrameni test va train setlarga ajratish
# DataFrameni narxlar qismini alohida dfga saqlab olish va int bo'lmagan qismni ajratib olish

from sklearn.model_selection import train_test_split
train_set, test_set = train_test_split(df_processed, test_size=0.2, random_state=42)

X_train = train_set.drop("price", axis=1)
y = train_set["price"].copy()

X_num = X_train.drop('district', axis=1)
```

```python
# Ajaratib olingan raqamli dfni sandart sonlarga aylantirish

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler

num_pipeline = Pipeline([
    ('std_scaler', StandardScaler())
])

num_pipeline.fit_transform(X_num)
```

```python
from sklearn.compose import ColumnTransformer

num_attribs = list(X_num)
cat_attribs = ['district']

full_pipeline = ColumnTransformer([
    ('num', num_pipeline, num_attribs),
    ('dist', OneHotEncoder(), cat_attribs)
])
```

```python
X_prepared = full_pipeline.fit_transform(X_train)
X_prepared
```

```python
### Liner Regression

from sklearn.linear_model import LinearRegression

LR_model = LinearRegression()
```

```python
LR_model.fit(X_prepared, y)
```

```python
test_data = X_train.sample(10)
test_data
```

```python
test_label = y.loc[test_data.index]
test_label
```

```python
test_data_prepared = full_pipeline.transform(test_data)
test_data_prepared
```

```python
test_data_prepared = full_pipeline.transform(test_data)
predicted_labels = LR_model.predict(test_data_prepared)
```

```python
from sklearn.metrics import mean_absolute_error

mae = mean_absolute_error(y_test, y_predicted)
print("MAE =", mae)
```

```python
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(y_test, y_predicted)
print("RMSE =", np.sqrt(mse))
```

```python
### Random Forest

from sklearn.ensemble import RandomForestRegressor
RF_model = RandomForestRegressor()
RF_model.fit(X_prepared, y)
```

```python
y_predicted = RF_model.predict(X_test_prepared)
```

```python
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(y_test, y_predicted)
print("RMSE =", np.sqrt(mse))
```

```python
# cross validation

X = df_processed.drop("price", axis=1)
y = df_processed["price"].copy()

X_prepared = full_pipeline.transform(X)
```

```python
from sklearn.model_selection import cross_val_score

mse_scores = cross_val_score(LR_model, X_prepared, y, scoring="neg_mean_squared_error", cv=5)

mse_scores
```

```python
def display_scores(scores):
  print("Scores:", scores),
  print("Mean:", scores.mean()),
  print("Std.dev:", scores.std())
```

```python
display_scores(np.sqrt(-mse_scores))
```

```python
scores = cross_val_score(RF_model, X_prepared, y, scoring="neg_mean_squared_error", cv=10)
LR_rmse_scores = np.sqrt(-scores)
display_scores(LR_rmse_scores)
```
